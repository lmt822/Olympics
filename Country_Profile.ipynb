```
def country_flag(df):
    if df['ISO'] in di.keys():
        return di[df['ISO']]
    else:
        return df['ISO']

country_cleaned['NOC']=country_cleaned.apply(country_flag, axis = 1)
```

```
country_cleaned = country_cleaned.drop(columns=['ISO'])
```

```
country_cleaned.head(5)
```

```
country_cleaned = country_cleaned.rename(columns={"country": "Country_name", "Surface area (km2)": "Area", "Population in thousands (2017)": "Population", "Sex ratio (m per 100 f, 2017)": "Sex_ratio", "GDP: Gross domestic product (million current US$)": "GDP", "Unemployment (% of labour force)": "UnE_rate"})
```

```
country_cleaned.shape
```

```
country_nonexist = pd.Series([np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan], index=['Country_name', 'Area', 'Population', 'Sex_ratio', 'GDP', 'UnE_rate', 'NOC'])
country_cleaned = country_cleaned.append(country_nonexist, ignore_index=True)

country_cleaned.loc[country_cleaned.NOC.isnull(),'NOC'] = 'ERR'
country_cleaned.loc[country_cleaned.Country_name.isnull(),'Country_name'] = 'No longer exist'

country_cleaned = country_cleaned.drop_duplicates()
```

```
country_cleaned = country_cleaned.replace('...', np.NaN)
country_cleaned = country_cleaned.replace('~0', '0')
```

```
country_cleaned['UnE_rate'] = country_cleaned['UnE_rate'].astype('float64')
country_cleaned['Area'] = country_cleaned['Area'].astype('float64')
```
